```python
import bt
import matplotlib.pyplot as plt
plt.style.use('seaborn')
%matplotlib inline
```

```python
def long_only(tickers, start='2010-01-01', name='long_only'):
    data = bt.get(tickers=tickers,start=start)
    s = bt.Strategy(name=name, 
                    algos=[bt.algos.RunOnce(),
                           bt.algos.SelectAll(),
                           bt.algos.WeighEqually(),
                           bt.algos.Rebalance()])
    return bt.Backtest(strategy=s, data=data)

port = long_only(['msft','aapl','tsla','nvda'], name='port')
big2 = long_only(['msft','aapl'], name='big2')
future2 = long_only(['nvda','tsla'], name='future2')
qqq = long_only(['qqq'], name='qqq')
spy = long_only(['spy'], name='spy')

res = bt.run(port, big2, future2, qqq, spy)
```

```python
res.plot(linewidth=0.85);
```
